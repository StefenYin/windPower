```python
def new_square(angle, side, rotorDiameter):    """Return a new square area for the farm layout.        Parameters    ----------    angle : float (radian)        270 minus the wind direction.    side : float (mile)        The side of the original square, 2 miles by 2 miles.    rotorDiameter : float (meter)        The diameter of the turbine rotor.    Return    ------    newside : float (meter)        The square side after subtracting setback.    X1 : float (meter)        The small distance in the transfered square according to the angle for wind farm.    X2 : float (meter)        The big distance in the transfered square according to the angle for wind farm.    Xt : float (meter)        The side length of the transfered square for wind farm.    """        D = rotorDiameter    mile_2_meter = 1609.34    ft_2_meter = 0.3048        setback = max(3*D, 1000*ft_2_meter)   #1000ft comparison for setback        newside = side*mile_2_meter - 2*setback        X1 = sin(angle) * newside    X2 = cos(angle) * newside    Xt = X1 + X2        return newside, X1, X2, Xt
```

```python
def numbers_and_remaining(length, distanceToBoundary, differenceTime, rotorDiameter):    """Return the numbers and remaining distance to the other boundary."""    numbers = int((length - distanceToBoundary)//(differenceTime * rotorDiameter)) + 1    #or use the floor func in numpy or math    remaining = (length - distanceToBoundary)%(differenceTime * rotorDiameter)        return numbers, remaining
```

```python
def rows_number_man(eachrow_distance_man, Xt):    """Return the rows that can be laied out in the limited farm when mannually inputting the distance        Parameter    ---------    eachrow_distance_man : list        A list of distance between each row. The first row distance is from the south west corner point.        [s0, s1, s2, s3, ...]        Return    ------    rows_man : integer        The number of rows for thoes input distance.    eachrow_distance : a list        The valid distance in the limited farm square.    remain_man : float        The remaining distance.    """    sum = 0.0    for no, distance in enumerate(eachrow_distance_man):        sum += distance        if sum > Xt:    #meet the limited side length            rows_man = no            remain_man = Xt - (sum-distance)            break    if sum < Xt:   #total summation        remain_man = Xt-sum        rows_man = no + 1    eachrow_distance = eachrow_distance_man[:rows_man]        return rows_man, eachrow_distance, remain_man
```

```python
def each_row(rows):    """    Return    ------    d_firstturbine : a list        A list of distances between first turbine and the boundary for each row,    c_eachrow : a list        A list of times of rotor diameter to define the distance between turbines in each row.    Note: Here, we assume the times is constant in each row."""        d_str = ['d'+str(number) for number in range(rows)]    c_str = ['c'+str(number) for number in range(rows)]        d_firstturbine = []    c_eachrow = []    while len(d_firstturbine)!=rows or len(c_eachrow)!=rows:        d_input = raw_input("Please input a list of values for the distance of the first turbine of each row,\n"                            "following the list: %s\n"%d_str)        c_input = raw_input("Please input a list of values for times distance of each row,\n"                            "following the list: %s\n"%c_str)        d_firstturbine= [float(input) for input in d_input.split()]        c_eachrow= [float(input) for input in c_input.split()]        return d_firstturbine, c_eachrow#You can assign the list of d and c.
```

```python
def each_row_information(angle, newSide, X1, X2, Xt, rows, rotorDiameter, s0, rowTimes, d_firstTurbine, c_eachRow):    """    Return    ------    eachrow_infor : a list        A list of tuples containing information of each row, in the following order:        e.g. the first row:         (length of a row(L), distance of first turthe(d), side times of a row(c), numbers of turbine in a row(N), remainning distance of a row(r))    """    D = rotorDiameter    d = d_firstTurbine    c = c_eachRow        row_infor = []        for i in range(rows):        if (rowTimes*D*i + s0) < X1:            L = Xt - tan(angle) * (X2-(rowTimes*D*i+s0)) - (X1 - (rowTimes*D*i+s0))/tan(angle)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)                    if (rowTimes*D*i + s0) >= X1 and (rowTimes*D*i + s0) <= X2:            L = newSide/cos(angle)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)                    if (rowTimes*D*i + s0) >= X2:            L = Xt - (rowTimes*D*i+s0 - X2)/tan(angle) - tan(angle) * (rowTimes*D*i+s0 - X1)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)    return row_infordef each_row_information_man(angle, newSide, X1, X2, Xt, rotorDiameter,                              eachrow_DistanceMan, d_firstTurbine, c_eachRow):    """    Return    ------    eachrow_infor_man : a list        A list of tuples containing information of each row which is mannully input.    """    D = rotorDiameter    d = d_firstTurbine    c = c_eachRow        row_infor = []        sum = 0.0    for i, value in enumerate(eachrow_DistanceMan):        sum += value        if sum < X1:            L = Xt - tan(angle) * (X2-sum) - (X1 - sum)/tan(angle)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)                    if sum >= X1 and sum <= X2:            L = newSide/cos(angle)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)                    if sum > X2:            L = Xt - (sum - X2)/tan(angle) - tan(angle) * (sum - X1)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)    return row_infor
```

```python
def each_turbine_position_man(angle, X2, rowsMan, rowInforMan, s, rotorDiameter):    """Return the position of every turbine along the corresponding coordinates about the farm.    Return    ------    each_row : a list        A list of each row containing the position of every turbine in the row, and the position is expressed by a tuple of        (X,Y) in the transfered coordinates.    Note    ----    Here, the coordinates here is built on the main wind direction for wind farm initial layout.    """    D = rotorDiameter    each_row = []    sum = 0.0    for i in range(rowsMan):        d = rowInforMan[i][1]        c = rowInforMan[i][2]        row = []        sum += s[i]        for j in range(rowInforMan[i][3]):            X = sum            if X <= X2:                Y = d + tan(angle) * (X2-X) + c*D*j            if X > X2:                Y = d + (X - X2)/tan(angle) + c*D*j                            position = (X, Y)            row.append(position)                each_row.append(row)    return each_row
```

```python
def each_turbine_position(angle, X2, rows, rowInfor, rowTimes, s0, rotorDiameter):    """Return the position of every turbine along the corresponding coordinates about the farm.    Return    ------    each_row : a list        A list of each row containing the position of every turbine in the row, and the position is expressed by a tuple of        (X,Y) in the transfered coordinates.    Note    ----    Here, the coordinates here is built on the main wind direction for wind farm initial layout.    """    D = rotorDiameter    each_row = []    for i in range(rows):        d = rowInfor[i][1]        c = rowInfor[i][2]        row = []        for j in range(rowInfor[i][3]):            X = s0 + rowTimes*D*i                        if (rowTimes*D*i + s0) <= X2:                Y = d + tan(angle) * (X2-(s0+rowTimes*D*i)) + c*D*j            if (rowTimes*D*i + s0) > X2:                Y = d + (rowTimes*D*i + s0 - X2)/tan(angle) + c*D*j                            position = (X, Y)            row.append(position)                each_row.append(row)    return each_row
```

```python
def transfered_new_position(rotating_angle, position):    """Return the position of every turbine in the farm relative to a new coordinates.        Parameter    ---------    rotating_angle : float (radian)        The angle between the main wind direction and other wind direction.        The positive rotation angle is counter-clockwise due to initial coordinates.    position : list        The position of every turbine in the initial coordinates, expressed in a tuple but        contained in a final list.    Return    ------    new_position : list        The new position of every turbine in a new coordinates.    """        cos_matrix = np.array([[cos(angle), sin(angle)],[-sin(angle), cos(angle)]])        new_position = []    position = np.array(position)        for row in range(len(position)):        new_position_stage = []        for po in position[row]:            X_new = sum((cos_matrix*po)[0])            Y_new = sum((cos_matrix*po)[1])            new_position_stage.append((X_new, Y_new))        new_position.append(new_position_stage)        return new_positiondef transfered_new_row(new_position):    """Return the sorted new position in each new row which is parallel to Y axis (or perpendicular to the wind direction).        Parameter    ---------    new_position : list        The position of every turbine in a new rotated coordinates.    Return    ------    newRow_position : list       The position of every turbine in the rotated coordinate, but sorted in each new row which is       perpendicular to the wind direction.    Note    ----    1, The same coordinate value may change a little since the last digit of a long float is random.    2, We still need to delete some rows since the difference between each new row is not the minimum (5*D) now.    """        position_ready = []    for i in range(len(new_position)):        for po in new_position[i]:            position_ready.append(po)    position_ready.sort()        newRow_position = []    stage2 = []        for no, po in enumerate(position_ready):        stage1 = []                if no <= (len(position_ready)-2) and abs(position_ready[no][0] - position_ready[no+1][0]) < 0.001:               stage2.append(po)            if no == (len(position_ready)-2):                stage2.append(position_ready[no+1])                stage2.sort(key=itemgetter(1))   #sort the tuples according to the Y coordinate value                newRow_position.append(stage2)                break        else:            if stage2 == []:                stage1.append(po)                newRow_position.append(stage1)            else:                stage2.append(po)                stage2.sort(key=itemgetter(1))                newRow_position.append(stage2)                stage2 = []                    return newRow_position
```

```python
################################################################################################### position in the corresponding coordinates  -->  overlap area  -->  velocity deficit --> velocity# from the final function, we know the variable "eachTurbinePosition" is built on a list containing# position of turbines in a row and then next row, and in the corrsponding coordinates.# So, if you change the coordinates, or change the position of the turbine in another way like redesign# the spacing, just make sure that the way that you express the position of the turbine is the same # as the "eachTurbinePosition".###################################################################################################def overlap_area(position0, position1, rotorDiameter):    """Return the overlap area of downwind circle and rotor sweaping area as well as the relative position along the     transfered coordinates.    Parameter    ---------    position0, position1: tuple        A tuple of X-Y coordinates of a turbine in the farm.    Return    ------    overlap : float        The overlap area of front turbine downstream area and behind turbine.    X_01 : float        The relative position along the X coordinates.    Y_01 : float        The relative position along the Y coordinates.    Note    ----    1, In our case, the rotor diameter is same for all turbines.    2, The coodinates where the position of the turbine belong  can be transfered to any other coordinates which is gonna depend on the         angle between wind direction and mainstream of wind direction (which is used for design the wind farm)    """    k = 0.1   #wake entrainment constant    D = rotorDiameter   #diameter of the other circle    X_01 = position1[0] - position0[0]  #relative position, the vector from turbine0 to turbine1.    r = D/2    if X_01 < 0:        raise ValueError("The second turbine(1) is in front of the first turbine(0).")    else:        Y_01 = abs(position1[1] - position0[1])   #two circle centers distance        R = D/2 + k*X_01      #the radius of one circle                if Y_01 >= (R + r):            overlap = 0                    if Y_01 < (R + r) and Y_01 >= sqrt(R**2-r**2):            X_11 = (R**2 - r**2 + Y_01**2)/(2*Y_01)            X_22 = (-R**2 + r**2 + Y_01**2)/(2*Y_01)            alpha = arccos(X_11/R); beta = arccos(X_22/r)            overlap = alpha*R**2 - X_11*sqrt(R**2-X_11**2) - X_22*sqrt(r**2-X_22**2) + beta*r**2                    if Y_01 < sqrt(R**2-r**2) and Y_01 > (R-r):            X_1 = (R**2 - r**2 - Y_01**2)/(2*Y_01)            X_2 = (R**2 - r**2 + Y_01**2)/(2*Y_01)            alpha = arccos(X_2/R); beta = arccos(X_1/r)            overlap = alpha*R**2 - X_2*sqrt(R**2-X_2**2) + X_1*sqrt(r**2-X_1**2) + (pi*r**2-beta*r**2)                    if Y_01 <= (R-r):            overlap = pi*r**2                return overlap, X_01, Y_01def velocity_deficit(V0, thrustCoefficient, rotorDiameter, overlapArea, X_01):    """Return the velocity deficit of turbine1 affected by the downwind from turbine0.    Parameter    ---------    V0 : float        The incident wind speed at turbine0.    thrustCoefficient : float        The thrust coefficient of turbine0, corresponding to the V0. Here we use a constant value.    Note    ----    Here, turbine1 and turbine0 have the same radius, rotorDiameter/2.    """    k = 0.1    Ct = thrustCoefficient    D = rotorDiameter        V01 = V0 * (1-sqrt(1-Ct)) * (D/(D+2*k*X_01)) * overlapArea / (pi*(D/2)**2)        return V01def velocity_and_deficit_a_row(windSpeed, velocityDeficitInRow):    """Return a list of velocities and velocity deficits of each turbine in a row affected by all turbines of front rows.        Parameter    ---------    windSpeed : float        The initial wind speed.    velocityDeficitInRow : an array-like, m-by-n        A list of deficit of every turbine in a row affected by all upstream turbines.        m : the number of turbines upstream; n : the number of turbines in a row.    """    VDinrow = velocityDeficitInRow    VD = []    for i in range(len(VDinrow[0])):        VD_specific_turbine_list = map(itemgetter(i),VDinrow)        sum = 0        for j in VD_specific_turbine_list:            sum += j**2        VD_total = sqrt(sum)        VD.append(VD_total)    V = [(windSpeed-value) for value in VD]    return V, VDdef total_velocity_and_deficit(windSpeed, eachTurbinePosition, rotorDiameter, thrustCoefficient):    """Return the total velocity and deficit of every turbine in the farm."""    D = rotorDiameter    TP = eachTurbinePosition    Ct = thrustCoefficient        V_all = []    VD_all = []        V_0 = [windSpeed for position in TP[0]]    VD_0 = [0.0 for position in TP[0]]    V_all.append(V_0)    VD_all.append(VD_0)        m_wake = [[] for i in range(len(TP))]        for no in range(len(TP))[:-1]:                for po_upstream, vel in zip(TP[no], V_all[no]):  #choose a row containing the position and velocity of every turbine            a_wake = [[] for j in range(len(TP))]                        for i in range(len(TP))[(no+1):]:   #choose the downstream rows                                for po_downstream in TP[i]:                    #thrust coeficient corresponding to the wind speed for each turbine in a upstream row. Here, we use a constant thrust coefficient for each turbine and each type of turbine                    overlap, X_updown, Y_updown = overlap_area(po_upstream, po_downstream, D)                    veldef = velocity_deficit(vel, Ct, D, overlap, X_updown)                    a_wake[i].append(veldef)                m_wake[i].append(a_wake[i])                        V_row, VD_row = velocity_and_deficit_a_row(windSpeed, m_wake[no+1])   #Calculate the complete downstream row        V_all.append(V_row)        VD_all.append(VD_row)    return V_all, VD_all
```
