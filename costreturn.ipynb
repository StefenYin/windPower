```python
############################################# total energy -> cost of energy, inital capital cost, return on investment, years to return.################################################def cost_of_energy(annualenergyproduction, ratedpower, rotordiameter, towerheight, numbersofturbines, typeturbine='turbine3'):    """Return cost of energy per kW.h at the specified wind farm.    Note:    The cost per kW also depend on what types of turbines are going to be used.    turbine1: fixed-pitch and constant rotor speed with 90% discount    turbine2: variable-pitch and constant rotor speed with 95% discount.    Default is turbine3: variable-pitch and variable rotor speed.    """    AEP = annualenergyproduction    rp = ratedpower   #rated power in MW    D = rotordiameter    H = towerheight    no = numbersofturbines        FCR = 0.1   #annual fixed charge rate    OM = 0.01   #0.01/kWh operation and Maintenance cost rate    disc1 = 0.90  #90% discount on windturbine capital cost    disc2 = 0.95  #95% discount         #InitialCapitalCost    RA = 126700 * (D/40.0)**2.53  #Rotor Assembly    DTN = 173118 * (rp/0.5)**0.76   #Drive Train Nacelle    Tower = 286882 * (H/80)**1.667  #Tower    EAC = 100000 * (rp/0.5)**0.76  #Electrical and Controls    WindTurbineCC = RA + DTN + Tower + EAC  #Wind Turbine Capital Cost. Here, depends on different types of turbine        if typeturbine == 'turbine1':        WindTurbineCC = disc1 * WindTurbineCC    if typeturbine == 'turbine2':        WindTurbineCC = disc2 * WindTurbineCC    BOPperturbine = 177150 * (rp/0.5)**0.76   #Balance of plant per turbine    BOP = BOPperturbine * no        ICC = BOP + WindTurbineCC * no  #InitialCapitalCost, Independent of annual energy production        COE = ICC * FCR / AEP + OM        return COE, ICC
```

```python
def return_on_investment(annualenergyproduction, years, ratedpower, rotordiameter,                          towerheight, numbersofturbines, typeturbine='turbine3', productiontaxcredit='Yes'):    """Return the benefits after investment."""        AEP = annualenergyproduction    rp = ratedpower   #rated power in MW    D = rotordiameter    H = towerheight    no = numbersofturbines    POE = 0.10 #price of energy 0.1/kWh        if productiontaxcredit == 'Yes':        PTC = 0.021    else:        PTC = 0.0    COE, ICC = cost_of_energy(AEP, rp, D, H, no, typeturbine = typeturbine)        total_EP = years * AEP        ROI = (POE - COE + PTC) * total_EP / ICC        return ROIdef return_years(annualenergyproduction, ratedpower, rotordiameter,                  towerheight, numbersofturbines, typeturbine='turbine3', productiontaxcredit='Yes'):    """Return the time to return on the investment."""        AEP = annualenergyproduction    rp = ratedpower   #rated power in MW    D = rotordiameter    H = towerheight    no = numbersofturbines    POE = 0.10 #price of energy 0.1/kWh        if productiontaxcredit == 'Yes':        PTC = 0.021    else:        PTC = 0.0    COE, ICC = cost_of_energy(AEP, rp, D, H, no, typeturbine = typeturbine)        years = ICC / ((POE-COE+PTC) * AEP)        if years > 10.0:        years_before = 10.0        years_after = (ICC - (POE-COE+PTC)*10.0*AEP) / ((POE-COE) * AEP)        years_all = years_before + years_after    else:        years_all = years        return years_all        
```
