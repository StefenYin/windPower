```python
########################################## original vel -> vel after deficits for energy calculation# one and two farm here are first two rows in front and adding two back rows.# And the second row of each two is not affected by the front row.######################################from numpy import sin, cos, tan, arccos, sqrt, piimport numpy as npfrom operator import itemgetter
```

```python
def new_square(angle, side, rotorDiameter):    """Return a new square area for the farm layout.        Parameters    ----------    angle : float (radian)        270 minus the wind direction.    side : float (mile)        The side of the original square, 2 miles by 2 miles.    rotorDiameter : float (meter)        The diameter of the turbine rotor.    Return    ------    newside : float (meter)        The square side after subtracting setback.    X1 : float (meter)        The small distance in the transfered square according to the angle for wind farm.    X2 : float (meter)        The big distance in the transfered square according to the angle for wind farm.    Xt : float (meter)        The side length of the transfered square for wind farm.    """        D = rotorDiameter    mile_2_meter = 1609.34    ft_2_meter = 0.3048        setback = max(3*D, 1000*ft_2_meter)   #1000ft comparison for setback        newside = side*mile_2_meter - 2*setback        X1 = sin(angle) * newside    X2 = cos(angle) * newside    Xt = X1 + X2        return newside, X1, X2, Xt
```

```python
def numbers_and_remaining(length, distanceToBoundary, differenceTime, rotorDiameter):    """Return the numbers and remaining distance to the other boundary."""    numbers = int((length - distanceToBoundary)//(differenceTime * rotorDiameter)) + 1    #or use the floor func in numpy or math    remaining = (length - distanceToBoundary)%(differenceTime * rotorDiameter)        return numbers, remaining
```

```python
def each_row_information_man(angle, newSide, X1, X2, Xt, rotorDiameter,                              eachrow_DistanceMan, d_firstTurbine, c_eachRow):    """    Return    ------    eachrow_infor_man : a list        A list of tuples containing information of each row which is mannully input.    (length of a row(L), distance of first turthe(d), side times of a row(c),     numbers of turbine in a row(N), remainning distance of a row(r))    """    D = rotorDiameter    d = d_firstTurbine    c = c_eachRow        row_infor = []        sum = 0.0    for i, value in enumerate(eachrow_DistanceMan):        sum += value        if sum < X1:            L = Xt - tan(angle) * (X2-sum) - (X1 - sum)/tan(angle)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)                    if sum >= X1 and sum <= X2:            L = newSide/cos(angle)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)                    if sum > X2:            L = Xt - (sum - X2)/tan(angle) - tan(angle) * (sum - X1)            N, r = numbers_and_remaining(L, d[i], c[i], D)            eachrow = (L, d[i], c[i], N, r)            row_infor.append(eachrow)    return row_infor
```

```python
def each_turbine_position_man(angle, X2, rowsMan, rowInforMan, s, rotorDiameter):    """Return the position of every turbine along the corresponding coordinates about the farm.    Return    ------    each_row : a list        A list of each row containing the position of every turbine in the row, and the position is expressed by a tuple of        (X,Y) in the transfered coordinates.    Note    ----    Here, the coordinates here is built on the main wind direction for wind farm initial layout.    """    D = rotorDiameter    each_row = []    sum = 0.0    for i in range(rowsMan):        d = rowInforMan[i][1]        c = rowInforMan[i][2]        row = []        sum += s[i]        for j in range(rowInforMan[i][3]):            X = sum            if X <= X2:                Y = d + tan(angle) * (X2-X) + c*D*j            if X > X2:                Y = d + (X - X2)/tan(angle) + c*D*j                            position = (X, Y)            row.append(position)                each_row.append(row)    return each_row
```

```python
def overlap_area(position0, position1, rotorDiameter):    """Return the overlap area of downwind circle and rotor sweaping area as well as the relative position along the     transfered coordinates.    Parameter    ---------    position0, position1: tuple        A tuple of X-Y coordinates of a turbine in the farm.    Return    ------    overlap : float        The overlap area of front turbine downstream area and behind turbine.    X_01 : float        The relative position along the X coordinates.    Y_01 : float        The relative position along the Y coordinates.    Note    ----    1, In our case, the rotor diameter is same for all turbines.    2, The coodinates where the position of the turbine belong  can be transfered to any other coordinates which is gonna depend on the         angle between wind direction and mainstream of wind direction (which is used for design the wind farm)    """    k = 0.1   #wake entrainment constant    D = rotorDiameter   #diameter of the other circle    X_01 = position1[0] - position0[0]  #relative position, the vector from turbine0 to turbine1.    r = D/2    if X_01 < 0:        raise ValueError("The second turbine(1) is in front of the first turbine(0).")    else:        Y_01 = abs(position1[1] - position0[1])   #two circle centers distance        R = D/2 + k*X_01      #the radius of one circle                if Y_01 >= (R + r):            overlap = 0                    if Y_01 < (R + r) and Y_01 >= sqrt(R**2-r**2):            X_11 = (R**2 - r**2 + Y_01**2)/(2*Y_01)            X_22 = (-R**2 + r**2 + Y_01**2)/(2*Y_01)            alpha = arccos(X_11/R); beta = arccos(X_22/r)            overlap = alpha*R**2 - X_11*sqrt(R**2-X_11**2) - X_22*sqrt(r**2-X_22**2) + beta*r**2                    if Y_01 < sqrt(R**2-r**2) and Y_01 > (R-r):            X_1 = (R**2 - r**2 - Y_01**2)/(2*Y_01)            X_2 = (R**2 - r**2 + Y_01**2)/(2*Y_01)            alpha = arccos(X_2/R); beta = arccos(X_1/r)            overlap = alpha*R**2 - X_2*sqrt(R**2-X_2**2) + X_1*sqrt(r**2-X_1**2) + (pi*r**2-beta*r**2)                    if Y_01 <= (R-r):            overlap = pi*r**2                return overlap, X_01, Y_01def velocity_deficit(V0, thrustCoefficient, rotorDiameter, overlapArea, X_01):    """Return the velocity deficit of turbine1 affected by the downwind from turbine0.    Parameter    ---------    V0 : float        The incident wind speed at turbine0.    thrustCoefficient : float        The thrust coefficient of turbine0, corresponding to the V0. Here we use a constant value.    Note    ----    Here, turbine1 and turbine0 have the same radius, rotorDiameter/2.    """    k = 0.1    Ct = thrustCoefficient    D = rotorDiameter        V01 = V0 * (1-sqrt(1-Ct)) * (D/(D+2*k*X_01)) * overlapArea / (pi*(D/2)**2)        return V01def velocity_and_deficit_a_row(windSpeed, velocityDeficitInRow):    """Return a list of velocities and velocity deficits of each turbine in a row affected by all turbines of front rows.        Parameter    ---------    windSpeed : float        The initial wind speed.    velocityDeficitInRow : an array-like, m-by-n        A list of deficit of every turbine in a row affected by all upstream turbines.        m : the number of turbines upstream; n : the number of turbines in a row.    """    VDinrow = velocityDeficitInRow    VD = []    for i in range(len(VDinrow[0])):        VD_specific_turbine_list = map(itemgetter(i),VDinrow)        sum = 0        for j in VD_specific_turbine_list:            sum += j**2        VD_total = sqrt(sum)        VD.append(VD_total)    V = [(windSpeed-value) for value in VD]    return V, VDdef total_velocity_and_deficit(windSpeed, eachTurbinePosition, rotorDiameter, thrustCoefficient):    """Return the total velocity and deficit of every turbine in the farm."""    D = rotorDiameter    TP = eachTurbinePosition    Ct = thrustCoefficient        V_all = []    VD_all = []        V_0 = [windSpeed for position in TP[0]]    VD_0 = [0.0 for position in TP[0]]    V_all.append(V_0)    VD_all.append(VD_0)        m_wake = [[] for i in range(len(TP))]        for no in range(len(TP))[:-1]:                for po_upstream, vel in zip(TP[no], V_all[no]):  #choose a row containing the position and velocity of every turbine            a_wake = [[] for j in range(len(TP))]                        for i in range(len(TP))[(no+1):]:   #choose the downstream rows                                for po_downstream in TP[i]:                    #thrust coeficient corresponding to the wind speed for each turbine in a upstream row. Here, we use a constant thrust coefficient for each turbine and each type of turbine                    overlap, X_updown, Y_updown = overlap_area(po_upstream, po_downstream, D)                    veldef = velocity_deficit(vel, Ct, D, overlap, X_updown)                    a_wake[i].append(veldef)                m_wake[i].append(a_wake[i])                        V_row, VD_row = velocity_and_deficit_a_row(windSpeed, m_wake[no+1])   #Calculate the complete downstream row        V_all.append(V_row)        VD_all.append(VD_row)    return V_all, VD_all
```
