```python
################################################## define a power calculation for variable-speed################################################def power_varied_pitch(rotorDiameter, rotorSpeed, windSpeed, powerCoefficient, airDensity):    """Return the power output of variable-pitch condition.    Parameter    ---------    powerCoefficient : a list        The max power coefficient corresponding optimum pitch angle and tip speed ratio.    windSpeed : a list        The wind speed derived from tip speed ratio. So the length is same as powerCoefficient.    """        D = rotorDiameter    w = rotorSpeed    Cp = powerCoefficient    Vw = windSpeed    power = Cp*1.0/2*airDensity*Vw**3*pi*(D/2)**2/1000 #kw    return powerdef power_coefficient_varied_pitch(power, rotorDiameter, rotorSpeed, windSpeed, airDensity):    """Return the power coefficient when power is over one MW.    Note: the unit of power input here is kw."""    D = rotorDiameter    w = rotorSpeed    Vw = windSpeed    power_coefficient = power*1000/(1.0/2*airDensity*Vw**3*pi*(D/2)**2)    return power_coefficientpower_varied_pitch_vec = np.vectorize(power_varied_pitch)
```
